## Gaussian Naive Bayes
Talvez o mais simples classificador Naive Bayes seja o Gaussiano. Nesse classificador, a suposição é que os dados de cada rótulo são extraídos de uma distribuição gaussiana simples.

![Gauss](figuras/gauds.gif)

**Importando as bibliotecas principais**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

**Gerando os dados com distribuição gaussiana**

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='RdBu');
```

**Treinando o modelo**

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y);
```

**Gerando uma nova massa de dados**

```python
# Gerando os dados
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)

#Plotando
plt.scatter(Xnew[:, 0], Xnew[:, 1], c='green', alpha=0.2);
```

**Plotando os novos dados junto aos antigos**

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu') # Plot dos dados utilizados para treino
plt.scatter(Xnew[:, 0], Xnew[:, 1], c='green', alpha=0.2); # Plot dos novos dados ainda sem classificação
```

**Predizendo os novos valores com o modelo treinado**

```python
ynew = model.predict(Xnew)
```

**Plotando os novos valores**  
Perceba, a seguir, que os novos dados serão tematizados a partir da classificação feita pelo modelo.

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')  # Plot dos dados utilizados para treino
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.3);  # Plot dos novos dados classificados
```
