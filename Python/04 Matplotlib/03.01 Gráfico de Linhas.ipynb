## Gráfico de Linhas  
Gráfico de linhas é o tipo mais básico de plotagem. Esse gráfico é o padrão do matplotlib, para plotá-lo, utilizamos o ``plt.plot()``.

**Sintaxe:**
``plt.plot(índice, valores)``

**Nota:** Há outros argumentos que também podem ser inseridos, para a informação completa é melhor consultar a documentação.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.style.use('ggplot') # Escolhendo o estilo ggplot

ano = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013]
pibPcapitaBr = [8.72, 8.69, 8.81, 8.81, 9.22, 9.42, 9.69, 10.18, 10.58, 10.42, 11.07, 11.26, 11.27, 11.45]

plt.plot(ano, pibPcapitaBr)

plt.show()
```
