## Gráfico de Linhas  
Gráfico de linhas é o tipo mais básico de plotagem. Esse gráfico é o padrão do matplotlib, para plotá-lo, utilizamos o ``plt.plot()``.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.style.use('ggplot') # Escolhendo o estilo ggplot

dados = np.random.randn(50).cumsum() # Criando os dados

fig = plt.figure()  # Criando uma figure

fig = plt.plot(dados) # Plotando os dados dentro da figure
```
