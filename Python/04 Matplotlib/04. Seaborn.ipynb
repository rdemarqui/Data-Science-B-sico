## Seaborn
O Matplotlib antecedeu o Pandas por mais de uma década e, portanto, não foi projetado para utilizar os DataFrames do Pandas. Para visualizar esse tipo de dados, você deve extrair cada Série e, com frequência, concatená-los no formato correto. Para suprir essa deficiência do Matplotlib, foi criado o **Seaborn**.

O Seaborn é uma biblioteca Python para visualização de dados baseada no Matplotlib, desenhada para funcionar muito bem trabalhando em conjunto com objetos do tipo Pandas DataFrame. Fornece uma interface de alto nível para produção de gráficos com estilos de design excepcionais por padrão.

Galeria: http://seaborn.pydata.org/examples/index.html

**Importando o Seaborn**  
Por padrão importamos o Seaborn como ``sns``. Ao importarmos o Seaborn e utilizarmos o comando ``sns.set()`` os padrões de cores do Matplotlib serão trocados para o padrão Seaborn.

```python
import seaborn as sns
sns.set()
```

Vamos a seguir explorar alguns recursos gráficos do Seaborn, lembrando que também é possível gerá-los no Matplotlib, porém com o Seaborn a tarefa será simplificada.  
* distplot - histograma
* jointplot - relação entre duas variáveis, apresentando o coeficiente de Pearson
* pairplot - Correlaciona múltiplas váriáveis de um dataset utilizando gráficos facetados
* factorplot - Possibilidade de plotar point, bar, count, box, violin e strip de maneira facetada
* lmplot - regressão com a opção de plotar em facetas
* heatmap - mapa de calor

```python
# Carregando mais algumas bibliotecas para nos auxiliar
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
# Carregando alguns datasets
iris = sns.load_dataset("iris")
gorjetas = sns.load_dataset('tips')
planets = sns.load_dataset('planets')
```

### distplot

```python
# Plotando um hitograma com o Seaborn.
#As linhas que acompanham a distribuição chamam-se estimativa de densidade kernel(sigla em inglês: kde).

sns.distplot(iris['sepal_length'])
sns.distplot(iris['sepal_width']);
```

### jointplot  
Mostra a relação entre duas variáveis, apresentando o coeficiente de Pearson que varia de -1 a 1.  
Quanto mais próximo de 1, maior é a correlação positiva entre elas; -> "Quando uma aumenta a outra também aumenta"  
Quanto mais próximo de -1, maior é a correlação negativa entre elas; -> "Quando uma aumenta a outra diminui"  
Quanto mais próximo de 0, menor é a correlação positiva entre elas.

Por padrão, o tipo de gráfico do jointplot é de dispersão ('scatter'), mas temos outras opções, como ``'kde'``, ``'reg'``, ``'resid'`` e ``'hex'``.

```python
# Plotando o jointplot
sns.jointplot("total_bill", "tip", data=gorjetas, kind='reg');
```

### pairplot  
Correlaciona múltiplas váriáveis de um dataset

```python
sns.pairplot(iris, hue='species', size=2.5);
```

### factorplot
Uma maneira de visualizar dados com muitas variáveis categóricas é usar uma grade de facetas. Seaborn tem a função ``factorplot()`` que simplifica a criação de muitos tipos de gráficos facetados.  
O gráfico padrão é mostrado em pontos, mas outros gráficos podem ser escolhidos com o parâmetro `` kind``, incluindo boxplot, gráficos de violinos, gráficos de barra ou pontos.

kind : {``point``, ``bar``, ``count``, ``box``, ``violin``, ``strip``}

```python
sns.factorplot(x='day', y='total_bill', hue='time', col='smoker', kind='bar', data=gorjetas);
```

```python
# Factorplot tipo count 
sns.factorplot("year", data=planets, aspect=4.0, kind='count', hue='method', order=range(2001, 2015));
```

```python
# Factorplot tipo violoin
sns.factorplot("day", "total_bill", data=gorjetas, kind='violin');
```

```python
# Factorplot tipo boxplot, usando estilo white
with sns.axes_style(style='white'):
    g = sns.factorplot("day", "total_bill", "sex", data=gorjetas, kind="box")
```

### lmplot

```python
sns.lmplot(x="total_bill", y="tip", data=gorjetas, palette="Set1");
```

```python
#size: escolhe o tamanho do gráfico

sns.lmplot(x="total_bill", y="tip", row="sex", col="time", data=gorjetas, size=3);
```

### heatmap

```python
sns.heatmap(iris.corr(), annot= True, cmap='magma');
```
