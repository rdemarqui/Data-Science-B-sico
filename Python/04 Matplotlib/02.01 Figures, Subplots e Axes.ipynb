### Figures, Subplots e Axes  

**Figures**  
Os gráficos do Matplotlib residem dentro de um objeto Figure. No Matplotlib, a figura pode ser vista como uma tela, na qual todos os componentes de desenho serão plotados eixos, gráficos, texto e rótulos. Para criar um objeto figure basta atribuir ``plt.figure()`` à uma variável, porém no Jupyter Notebook nada aparecerá até que criemos algo dentro dessa ``figure``. 

**Subplots**  
Os subplots são divisões dentro da figure onde podemos, por exemplo, gerar 2 ou mais gráficos gráficos.

**Axes**  
Axes seriam os eixos que representam cada divisão do subplot.

![](subdivision.png)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()  # Criando uma figure

# Criando subplots dentro da figure
ax1 = fig.add_subplot(2, 2, 1) # Subplot 2x2, 1° quadrante
ax2 = fig.add_subplot(2, 2, 2) # Subplot 2x2, 2° quadrante
ax3 = fig.add_subplot(2, 2, 3) # Subplot 2x2, 3° quadrante

# Ao criarmos um plot, o Matplotlib o atribuirá ao último quadrante criado:
plt.plot(np.random.randn(50).cumsum(), 'k--')

# Porém podemos atribuir um gráfico para cada subplot. Para isso basta chamar a instância criada
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30));
```

Embora essa interface seja rápida e conveniente para plotagens simples, é fácil encontrar problemas. Felizmente, existe uma maneira melhor.

### Método orientado a objetos  
A interface orientada a objetos está disponível para essas situações mais complicadas e para quando você quer mais controle sobre sua figura.

**plt.subplots**  
Como é comum contruirmos gráficos com subplots, o matplotlib possui um metodo para facilitar essa construção, chamado ``plt.subplots()``, onde podemos mapear cada subplot utilizando sua posição [linha, coluna] e atribuir um gráfico.

```python
# Criando um subplot 2x2
fig, axes = plt.subplots(2, 2)

# Os axes (eixos) são mapeados da seguinte forma:
axes[0, 0].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)         # Linha 0, coluna 0
axes[1, 0].plot(np.random.randn(50).cumsum())                                # Linha 1, coluna 0
axes[1, 1].scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))   # Linha 1, coluna 1


# Ajustando o espaçamento entre os subplots
plt.subplots_adjust(wspace=.4, hspace=.4)
```
