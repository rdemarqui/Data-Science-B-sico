## Gráfico de Pizza  
O gráfico de pizza é bom em casos que não necessitem contar uma evolução histórica e que não haja muita subdivisão, pois caso contrário ficará poluído. Valores muito desproporcionais também tentem a empilhar os dados tornando-o ilegível.  

**Sintaxe básica:**

    plt.pie(valores)

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

%matplotlib inline
```

Market share das 5 maiores operadora de telefonia móvel em 2017.  
Fonte: http://www.teleco.com.br/mshare.asp

```python
# Criando os dados
operadora = ['Vivo', 'Claro', 'TIM', 'Oi', 'Nextel']
celulares = [74940, 59022, 58634, 38942, 2788]

# Para o gráfico não ficar oval, configuramos seu tamanho
plt.figure(figsize=(5,5))

# Construindo o plot
plt.pie(celulares, labels=operadora, startangle=90, autopct='%.1f%%') # autopct mostra os percentuais no gráfico

# Melhorando a estética
plt.title('Market Share Operadoras Móveis em 2017') # Título do gráfico

plt.show()
```
