## Scatter Plot  
Gráficos Scatter Plot ajudam a verificar a correlação entre uma ou mais variáveis, inclusive destacando valores outliers.

**Sintaxe básica:**  

    plt.scatter(valor_X, valor_Y)

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

%matplotlib inline
```

No exemplo a seguir, faremos um scatter utilizando os dados da Heritage Foundation correlacionando o indice de liberdade econômica com o PIB per capita (dados 2018).   
Fonte: https://www.heritage.org/index/explore?view=by-variables

```python
# Utilizando os dados dos países das Américas
scoreLibEconAmer = [52.3, 63.3, 57, 57.1, 44.1, 51.4, 77.7, 75.2, 68.9, 65.6, 31.9, 
                        64.5, 61.6, 48.5, 63.2, 63.4, 58.7, 55.8, 60.6, 69.1, 64.8, 58.9, 
                        67, 62.1, 68.7, 67.6, 67.7, 48.1, 57.7, 75.7, 69.2, 25.2]

pibPerCapitaAmer = [20047, 24555, 17100, 8220, 7218, 15242, 46437, 24113, 14130, 
                        16436, 12390, 11375, 16049, 11109, 8909, 7899, 7873, 1784, 5271, 
                        8976, 18938, 5452, 23024, 9396, 12903, 11783, 11271, 13988, 31870, 
                        57436, 21527, 13761]

# Construindo o plot
plt.scatter(pibPerCapitaAmer, scoreLibEconAmer)

# Melhorando a estética
plt.xlabel('PIB Percapita U$') # Rótulo eixo x
plt.ylabel('Score Liberdade Economica') # Rótilo eixo y
plt.title('Liberdade x PIB') # Título do gráfico

plt.show()
```

Podemos inserir mais de um conjunto de dados, comparando o comportamento do grupo.

```python
# Plotando em conjunto com os países da Ásia e Pacífico
scoreLibEconAsiaPacif = [51.3, 80.9, 64.3, 55.1, 61.8, 53.9, 58.7, 57.8, 62, 90.2, 
                       54.5, 64.2, 72.3, 69.1, 50.8, 73.8, 62.8, 53.6, 70.9, 74.5, 
                       51.1, 52.3, 55.7, 54.1, 84.2, 54.4, 55.7, 65, 61.5, 88.8, 57.5, 
                       57.8, 76.6, 58.3, 67.1, 48.1, 63.1, 47.1, 51.5, 69.5, 53.1, 64.2]

pibPerCapitaAsiaPacif = [1919, 48899, 17439, 3891, 8227, 5832, 3737, 15399, 9268, 58322, 
                         6616, 11720, 41275, 25145, 1823, 37740, 3521, 5710, 95151, 27267, 
                         15553, 3234, 12275, 2479, 37294, 5106, 3541, 7728, 5553, 87855, 1973, 
                         12262, 48095, 3008, 16888, 4187, 5386, 17485, 6563, 2631, 6429, 76884]

# Construindo o plot
plt.scatter(pibPerCapitaAmer, scoreLibEconAmer)
plt.scatter(pibPerCapitaAsiaPacif, scoreLibEconAsiaPacif)

# Melhorando a estética
plt.xlabel('PIB Percapita U$') # Rótulo eixo x
plt.ylabel('Score Liberdade Economica') # Rótilo eixo y
plt.title('Liberdade x PIB') # Título do gráfico

plt.show()
```

Podemos verificar acima que uma maior liberdade econômica possui correlação positiva com alto PIB per capita, independente do continete analisado.
