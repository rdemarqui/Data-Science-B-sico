### Figures, Subplots e Axes  

**Figures**  
Os gráficos do Matplotlib residem dentro de um objeto Figure. No Matplotlib, a figura pode ser considerada como um único contêiner que contém todos os objetos que representam eixos, gráficos, texto e rótulos. Para criar um objeto figure basta atribuir ``plt.figure()`` à uma variável, porém no Jupyter Notebook nada aparecerá até que criemos algo dentro da ``figure``.  

**Subplots**  
Os subplots são divisões dentro da figure onde podemos, por exemplo, gerar 2 ou mais gráficos gráficos. Para gerar um subplot usamos ``.add_subplot()``.

**Axes**  
Axes seriam os quadrantes ou eixos que representam cada divisão do subplot.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()  # Criando uma figure

 # Criando subplots dentro da figure
ax1 = fig.add_subplot(2, 2, 1) # Subplot 2x2, 1° quadrante
ax2 = fig.add_subplot(2, 2, 2) # Subplot 2x2, 2° quadrante
ax3 = fig.add_subplot(2, 2, 3) # Subplot 2x2, 3° quadrante

# Ao criarmos um plot, o Matplotlib o atribuirá ao último quadrante criado:
plt.plot(np.random.randn(50).cumsum(), 'k--')

# Porém podemos atribuir um gráfico para cada subplot. Para isso basta chamar a instância criada
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

**plt.subplots**  
Como é comum contruirmos gráficos com subplots, o matplotlib possui um metodo para facilitar essa construção, chamado ``plt.subplots()``.

```python
# Criando um subplot 2x2
fig, axes = plt.subplots(2, 2)

# Os quadrantes (axes, ou eixos) são mapeados da seguinte forma:
axes[0, 0].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
axes[1, 1].scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))

# Ajustando o espaçamento entre os subplots
plt.subplots_adjust(wspace=.4, hspace=.4)
```
