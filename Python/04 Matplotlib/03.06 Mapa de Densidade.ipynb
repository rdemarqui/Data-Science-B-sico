## Mapa de densidade  

O histograma bi-dimensional no permite construir mapas de densidade, que é também uma forma de verificar o comportamento de um conjunto de dados em três dimensões.

**Histograma bi-dimensional**

**Sintaxe**

    plt.hist2d(valor_x, valor_y, bins, cmap)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Gerando os dados
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

**Hexagonal binnigs**  
Em vez de plotarmos as divisões (bins) quadradas, podemos também plotá-las em formato hexagonais. Para isso utilizamos a função ``plt.hexbin()``. Sua sintaxe é a mesma do histograma 2d.

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```
